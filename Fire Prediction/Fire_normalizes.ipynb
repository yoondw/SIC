```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
temp = pd.read_csv('fire_no_weather.csv', encoding = 'cp949').drop('Unnamed: 0', axis = 1)
temp
```

```python
from sklearn.preprocessing import RobustScaler

features = temp.iloc[:, 1:]
colnames = features.columns

scaler = RobustScaler()

features_scaled = pd.DataFrame(scaler.fit_transform(features), columns = colnames)
features_scaled
```

```python
final = pd.concat([temp.iloc[:, 0], features_scaled], axis = 1)
final
```

```python
final.to_csv('fire_robust.csv', encoding = 'cp949')
```

```python
temp2 = pd.read_csv('fire_no_weather.csv', encoding = 'cp949').drop('Unnamed: 0', axis = 1)
temp2
```

```python
fire_no = temp2[temp2['fr_yn'] == 0]
fire_yes = temp2[temp2['fr_yn'] == 1]
print(len(fire_no), len(fire_yes))
fire_no
```

```python
fire_no_selected = fire_no.sample(n = len(fire_yes), random_state = 54321)
fire_no_selected
```

```python
fire_1to1 = pd.concat([fire_no_selected, fire_yes], axis = 0).reset_index(drop = True)
fire_1to1
```

```python
fire_1to1.to_csv('fire_no_weather_1to1.csv', encoding = 'cp949')
```

```python
from sklearn.preprocessing import RobustScaler

features = fire_1to1.iloc[:, 1:]
colnames = features.columns

scaler = RobustScaler()

features_scaled = pd.DataFrame(scaler.fit_transform(features), columns = colnames)
features_scaled
```

```python
final = pd.concat([fire_1to1.iloc[:, 0], features_scaled], axis = 1)
final
```

```python
final.to_csv('fire_robust_1to1.csv', encoding = 'cp949')
```
